## 第一章 行列式 (Determinants)

行列式是线性代数的“计算器”，冲刺阶段重点在于**数值型的快速计算**和**抽象型的公式运用**。

### 1. 核心定义与性质

*   **本质**：行列式是一个**数**。
*   **展开定理**：
    *   行列式等于某一行（列）元素与其代数余子式乘积之和：$D = \sum_{k=1}^n a_{ik}A_{ik}$。
    *   **重要推论**：某一行（列）元素与另一行（列）对应元素的代数余子式乘积之和为0。
        $$ \sum_{k=1}^n a_{ik}A_{jk} = \begin{cases} D, & i=j \\ 0, & i \neq j \end{cases} $$
*   **重要性质**：
    1.  $|A^T| = |A|$。
    2.  互换两行（列），行列式变号。
    3.  某一行（列）有公因子 $k$，可提 $k$ 到行列式外面。
    4.  $|kA| = k^n |A|$ (**高频易错点**：$n$ 为矩阵阶数)。
    5.  两行（列）成比例，行列式为 0。
    6.  **倍加变换**：某一行（列）的 $k$ 倍加到另一行（列），行列式**值不变**（计算核心工具）。

### 2. 数值型行列式计算技巧

遇到具体数字的行列式，按以下优先级观察：

#### (1) 特殊结构法
*   **上/下三角形**：主对角线元素之积。
*   **副对角三角形**：$(-1)^{\frac{n(n-1)}{2}} \times$ 副对角线元素之积。
*   **拉普拉斯分块 (Laplace)**：
    $$
    \begin{vmatrix} A & 0 \\ C & B \end{vmatrix} = \begin{vmatrix} A & C \\ 0 & B \end{vmatrix} = |A| \cdot |B| \quad (\text{主对角块})
    $$
    $$
    \begin{vmatrix} 0 & A \\ B & C \end{vmatrix} = \begin{vmatrix} C & A \\ B & 0 \end{vmatrix} = (-1)^{mn} |A| \cdot |B| \quad (\text{副对角块}, A: m阶, B: n阶)
    $$

#### (2) 爪形（箭形）行列式
*   **特征**：第一行（列）非零，主对角线非零，其余全为0；或四边有数，中间空。
*   **解法**：利用对角线元素消去“爪尖”或“爪边”，化为三角行列式。

#### (3) 行（列）和相等
*   **特征**：每一行（列）元素之和都等于常数 $k$。
*   **解法**：将所有列加到第 1 列 $\to$ 提取公因子 $k$ $\to$ 第 1 列全为 1 $\to$ 此后每行减去第一行化0。

#### (4) 范德蒙德行列式 (Vandermonde)
*   **公式**：
    $$ D_n = \prod_{1 \le j < i \le n} (x_i - x_j) $$
*   **注意**：必须严格符合 $1, x_i, x_i^2, \dots, x_i^{n-1}$ 的结构。若缺项或次序不对，需补项或交换行列。

#### (5) 递推法 (异爪形/三对角)
*   适用于 $a_{ij}$ 有明显规律的 $n$ 阶行列式。
*   **思路**：按第一行（列）展开，建立 $D_n$ 与 $D_{n-1}, D_{n-2}$ 的关系，通常通过解特征方程或数学归纳法求解。

### 3. 抽象型行列式计算 (常用公式)

此类题型常结合矩阵运算考察：

1.  $|AB| = |A||B|$
2.  $|A^{-1}| = |A|^{-1}$
3.  $|A^*| = |A|^{n-1}$
4.  $|P^{-1}AP| = |A|$ (相似矩阵行列式相等)
5.  $|C^T A C| = |C|^2 |A|$ (合同)
6.  若 $A$ 的特征值为 $\lambda_1, \dots, \lambda_n$，则 $|A| = \prod_{i=1}^n \lambda_i$。

## 第二章 矩阵 (Matrices)

矩阵是线代的“容器”，重点是**运算规则**、**秩**以及**伴随矩阵**。

### 1. 矩阵运算陷阱

*   **乘法无交换律**：一般 $AB \neq BA$。
    *   推论：$(A+B)^2 = A^2 + AB + BA + B^2 \neq A^2 + 2AB + B^2$（除非 $A,B$ 可交换）。
    *   推论：$(AB)^k \neq A^k B^k$。
*   **消去律不成立**：$AB=AC, A \neq O \nRightarrow B=C$（除非 $A$ 可逆）。
*   **零因子**：$AB=O \nRightarrow A=O$ 或 $B=O$。

### 2. 伴随矩阵 ($A^*$) 全总结

*   **定义**：由代数余子式 $A_{ij}$ 构成的矩阵的**转置**。
*   **基本公式**：
    $$ AA^* = A^*A = |A|E $$
*   **常用性质** (设 $A$ 为 $n$ 阶可逆方阵)：
    1.  $|A^*| = |A|^{n-1}$
    2.  $(kA)^* = k^{n-1} A^*$ (**易错**)
    3.  $(A^*)^* = |A|^{n-2} A$
    4.  $(A^*)^{-1} = (A^{-1})^* = \frac{1}{|A|}A$
    5.  $(AB)^* = B^* A^*$ (注意次序颠倒)

### 3. 逆矩阵 ($A^{-1}$)

*   **存在条件**：$|A| \neq 0 \iff r(A)=n \iff$ 行/列向量线性无关 $\iff 0$ 不是特征值。
*   **求解方法**：
    1.  **具体型**：初等行变换 $(A|E) \xrightarrow{r} (E|A^{-1})$。
    2.  **公式法**：$A^{-1} = \frac{1}{|A|}A^*$ (适用于二阶或理论推导)。
    3.  **分块矩阵**：
        $$ \begin{pmatrix} A & 0 \\ 0 & B \end{pmatrix}^{-1} = \begin{pmatrix} A^{-1} & 0 \\ 0 & B^{-1} \end{pmatrix} $$
    4.  **多项式法**：若 $A^2 - 3A + 2E = O \implies A(A-3E) = -2E \implies A [-\frac{1}{2}(A-3E)] = E$。

### 4. 矩阵的秩 (Rank) —— 核心中的核心

**秩是线性代数的灵魂，连接方程组、向量组和矩阵。**

#### (1) 定义与计算
*   $r(A) = k \iff$ 最高阶非零子式为 $k$ 阶。
*   **计算**：一般通过初等行变换化为**行阶梯形**，非零行的行数即为秩。

#### (2) 重要不等式 (考研数学一重点)
1.  $0 \le r(A) \le \min(m, n)$
2.  $r(A) = r(A^T) = r(AA^T) = r(A^T A)$
3.  $r(A+B) \le r(A) + r(B)$
4.  $r(AB) \le \min(r(A), r(B))$
5.  **Sylvester不等式**：若 $A_{m \times n}, B_{n \times s}$，则 $r(AB) \ge r(A) + r(B) - n$。
6.  **零乘积结论**：若 $AB=O$，则 $r(A) + r(B) \le n$ (其中 $n$ 为 $A$ 的列数，$B$ 的行数)。

#### (3) 伴随矩阵的秩 (必背)
$$
r(A^*) = \begin{cases} n, & r(A)=n \quad (|A| \neq 0) \\ 1, & r(A)=n-1 \\ 0, & r(A)<n-1 \end{cases}
$$

### 5. 初等变换与初等矩阵

*   **法则**：
    *   **左乘**初等矩阵 = 进行对应**行**变换。
    *   **右乘**初等矩阵 = 进行对应**列**变换。
    *   $P_1 P_2 A Q_1 = B$ $\implies A, B$ **等价** ($r(A)=r(B)$)。

### 🎯 第一部分：典型“母题”演练 (行列式 & 矩阵)

#### 【母题 1】 抽象行列式与伴随矩阵 (2020/2022 类似考法)
设 $A$ 为 $3$ 阶可逆矩阵，且 $|2A^{-1} - 5A^*| = -12$，求 $|A|$。

> **思路点拨**：
> 看到 $A^*$ 和 $A^{-1}$ 同时出现，立刻使用公式 **$A^* = |A|A^{-1}$** 统一形式。切忌将 $A^*$ 展开或蛮算。

**解析**：
1.  代入公式：原式 $= |2A^{-1} - 5|A|A^{-1}| = |(2 - 5|A|)A^{-1}|$。
2.  利用 $|kA| = k^n|A|$：这里 $n=3$，系数 $k = (2-5|A|)$。
    $$ (2-5|A|)^3 \cdot |A^{-1}| = -12 $$
3.  利用 $|A^{-1}| = \frac{1}{|A|}$：
    $$ \frac{(2-5|A|)^3}{|A|} = -12 $$
4.  令 $x = |A|$，解方程 $\frac{(2-5x)^3}{x} = -12$。观察或整理可得 $x = \frac{2}{3}$ 或其他解（此处考察代数变形能力）。
    *(注：考研真题中数据通常凑得很整，如 $|A|=2$ 等，关键是第一步的统一变量)*

#### 【母题 2】 秩的逻辑判断 (选择题高频)
设 $A$ 是 $m \times n$ 矩阵，$B$ 是 $n \times m$ 矩阵，且 $AB=E_m$ (单位阵)。以下结论正确的是：
A. $m > n$
B. $r(A) = m$
C. $r(B) = n$
D. $BA = E_n$

> **思路点拨**：
> 这是一个秩的不等式陷阱题。利用 $r(AB) \le \min(r(A), r(B))$。

**解析**：
1.  因为 $AB=E_m$，两边取秩：$r(AB) = r(E_m) = m$。
2.  根据秩的不等式：$m = r(AB) \le \min(r(A), r(B))$。
3.  又因为 $A$ 是 $m \times n$，故 $r(A) \le m$ 且 $r(A) \le n$。
4.  结合 $r(A) \le m$ 和 $r(A) \ge m$，得出 **$r(A) = m$** (选 B)。
5.  同理 $r(B) \ge m$。由于 $B$ 是 $n \times m$，故 $r(B) \le m$，所以 $r(B)=m$。
6.  由此推导出 $m \le n$ (否则 $r(A)$ 不可能达到 $m$)。若 $m < n$，则 $A$ 矮胖，$B$ 瘦高，$BA$ 是 $n \times n$ 矩阵但秩只有 $m (<n)$，故 $BA$ 不可逆，不可等于 $E_n$。

#### 【母题 3】 矩阵方程与不可逆 (计算大题预备)
设 $A = \begin{pmatrix} 1 & 0 & 1 \\ 0 & 2 & 0 \\ 1 & 0 & 1 \end{pmatrix}$，且 $AP = P \Lambda$，其中 $\Lambda$ 为对角阵，$P$ 为可逆矩阵。求 $X$ 使得 $AX + X = A^2 + E$。

> **思路点拨**：
> 1. $AP=P\Lambda$ 暗示 $A$ 可相似对角化（这是第三部分的知识，但此处仅作为矩阵性质）。
> 2. 重点是解矩阵方程 $AX + X = A^2 + E$。
> 3. **千万别把 $A$ 拆开代入算！** 先移项分解。

**解析**：
1.  整理方程：$(A+E)X = A^2 + E$。
2.  欲求 $X$，需看 $A+E$ 是否可逆。
    $A+E = \begin{pmatrix} 2 & 0 & 1 \\ 0 & 3 & 0 \\ 1 & 0 & 2 \end{pmatrix}$。
    $|A+E| = 3(4-1) = 9 \neq 0$，故可逆。
3.  解法：$X = (A+E)^{-1}(A^2+E)$。
4.  **技巧**：这里不建议直接算 $(A+E)^{-1}$。观察右边 $A^2+E$ 能否因式分解？
    $A^2 + E = (A^2 - E) + 2E = (A+E)(A-E) + 2E$。
    所以 $X = (A+E)^{-1} [(A+E)(A-E) + 2E] = (A-E) + 2(A+E)^{-1}$。
    这样只需要算一次逆矩阵，且形式简单得多。

## 第三章 向量 (Vectors)

这一章的概念最容易晕，重点掌握**“相关性判定”**和**“极大无关组”**。

### 1. 线性相关与无关 (必考)

设向量组 $\alpha_1, \alpha_2, \dots, \alpha_s$：
*   **定义法**：$k_1\alpha_1 + \dots + k_s\alpha_s = 0$。
    *   若必有 $k_1=\dots=k_s=0 \implies$ **无关**。
    *   若存在不全为0的 $k \implies$ **相关**。

*   **三大判别准则 (秩、行列式、个数)**：
    1.  **行列式判别** ($n$ 个 $n$ 维向量)：
        $$ |\alpha_1, \dots, \alpha_n| \neq 0 \iff \text{线性无关} $$
    2.  **秩判别** (万能法)：
        向量组构成的矩阵 $A = (\alpha_1, \dots, \alpha_s)$。
        *   $r(A) = s$ (满秩，列数) $\iff$ **无关**。
        *   $r(A) < s \iff$ **相关**。
    3.  **个数判别**：
        *   若向量个数 $s >$ 维数 $n$，则**必相关** (矮胖矩阵 $Ax=0$ 必有非零解)。
        *   若向量组可由维数更低的向量组线性表示，则必相关。

*   **重要性质 (用于选择题推导)**：
    *   **部分与整体**：整体无关 $\implies$ 部分无关；部分相关 $\implies$ 整体相关。
    *   **延展性**：线性无关的向量组，添加分量（维数增加）后仍无关。线性相关的向量组，去掉分量（维数减少）后仍相关。（口诀：**无关组长高了还无关，相关组变矮了还相关**）。

### 2. 极大线性无关组与秩

*   **定义**：向量组的“骨架”，其余向量均可由其线性表示。
*   **求法** (重点)：
    1.  将向量 $\alpha_1, \dots, \alpha_s$ 拼成矩阵 $A = (\alpha_1, \dots, \alpha_s)$。
    2.  做**初等行变换**化为行阶梯形。
    3.  **台阶角（主元）所在的列**对应的**原向量**即为极大线性无关组。
    *   *注意：做行变换不改变列之间的线性关系，但不要取变换后的列向量，要取原向量。*

### 3. Schmidt 正交化 (数学一计算死穴)

在求正交基、实对称矩阵对角化时必用。
已知 $\alpha_1, \alpha_2, \alpha_3$ 线性无关，构造正交基 $\beta_1, \beta_2, \beta_3$：

1.  $\beta_1 = \alpha_1$
2.  $\beta_2 = \alpha_2 - \frac{(\alpha_2, \beta_1)}{(\beta_1, \beta_1)} \beta_1$
3.  $\beta_3 = \alpha_3 - \frac{(\alpha_3, \beta_1)}{(\beta_1, \beta_1)} \beta_1 - \frac{(\alpha_3, \beta_2)}{(\beta_2, \beta_2)} \beta_2$

*   **记忆口诀**：新向量 = 老向量 - (老向量在前面所有基上的投影)。
*   **最后一步**：单位化（标准化） $e_i = \frac{\beta_i}{||\beta_i||}$。

## 第四章 线性方程组 (Linear Equations)

这是把前面所有知识（行列式、矩阵、秩、向量）变现的章节。

### 1. 基础解系与通解 (齐次 $Ax=0$)

*   **有非零解条件**：$r(A) < n$ (未知数个数)。
    *   若是方阵：$|A|=0$。
*   **解的结构**：
    *   基础解系含 $n - r(A)$ 个解向量 ($\xi_1, \dots, \xi_{n-r}$).
    *   通解：$x = k_1\xi_1 + \dots + k_{n-r}\xi_{n-r}$。
*   **几何意义**：解空间是 $\mathbb{R}^n$ 的一个子空间，维数为 $n-r(A)$。

### 2. 非齐次方程组 ($Ax=b$)

*   **有解判定**：
    *   $r(A) \neq r(A|b) \implies$ 无解。
    *   $r(A) = r(A|b) = n \implies$ 唯一解。
    *   $r(A) = r(A|b) < n \implies$ 无穷多解。
*   **解的结构**：
    *   通解 = **齐次通解** + **特解**。
    *   $x = k_1\xi_1 + \dots + k_{n-r}\xi_{n-r} + \eta^*$。
*   **解的性质 (选择题常考)**：
    *   $\eta_1, \eta_2$ 是 $Ax=b$ 的解 $\implies \eta_1 - \eta_2$ 是 $Ax=0$ 的解。
    *   $\eta_1, \eta_2$ 是 $Ax=b$ 的解 $\implies \frac{1}{2}(\eta_1 + \eta_2)$ 也是解（仅当 $Ax=b$ 成立，实际上这要求 $b=0$ 或者系数和为1，此处更正：非齐次解的线性组合通常不是解，除非系数和为1。如 $c_1 \eta_1 + c_2 \eta_2$ 是解 $\iff c_1+c_2=1$）。

### 3. 三大高频考法总结

#### 考法一：含参方程组讨论 (大题No.1)
给定方程组（通常含参数 $a, \lambda$）。
1.  **写出增广矩阵** $(A|b)$。
2.  **初等行变换**化为阶梯形。
3.  根据台阶数讨论 $r(A)$ 和 $r(A|b)$ 的关系。
    *   **技巧**：若系数矩阵是方阵，优先算 $|A|$。
        *   $|A| \neq 0 \implies$ 唯一解（用克拉默法则或逆矩阵，但通常直接消元）。
        *   $|A| = 0 \implies$ 代入具体的参数值，回增广矩阵讨论无解或无穷多解。

#### 考法二：公共解问题 (Common Solutions)
求 $Ax=0$ 和 $Bx=0$ 的公共非零解。
*   **方法A (联立法)**：
    将两个方程组联立，构造 $\begin{pmatrix} A \\ B \end{pmatrix} x = 0$。求这个新方程组的基础解系。
*   **方法B (通解代入法)**：
    1. 先求 $Ax=0$ 的通解 $x = k_1\xi_1 + k_2\xi_2$。
    2. 将通解代入 $Bx=0$，即 $B(k_1\xi_1 + k_2\xi_2) = 0$。
    3. 得到关于 $k_1, k_2$ 的方程组，解出 $k_1, k_2$ 的关系。
    4. 代回原通解。

#### 考法三：同解方程组
已知 $Ax=0$ 和 $Bx=0$ 同解。
*   **结论**：同解 $\iff$ 解空间相同。
*   **推论**：若 $Ax=0$ 的解满足 $Bx=0$（即 $A$ 的解也是 $B$ 的解），则 $r(A) \ge r(B)$ 且 $B$ 的行向量可以由 $A$ 的行向量线性表示（即 $B = KA$）。
*   **做题技巧**：算出 $A$ 的通解，代入 $B$ 验证；或者利用秩的性质。

### 第三、四章 冲刺避坑指南

1.  **基础解系不唯一**：你求出的 $\xi$ 和答案不一样很正常，只要它们等价（可以互相线性表示）即可。
2.  **向量线性表示的顺序**：
    “$\beta$ 能由 $\alpha_1, \alpha_2$ 线性表示” $\implies$ 方程 $x_1\alpha_1 + x_2\alpha_2 = \beta$ 有解 $\implies r(\alpha_1, \alpha_2) = r(\alpha_1, \alpha_2, \beta)$。
    **切记**：$\beta$ 是被表示的，放在增广矩阵的最右边。
3.  **非齐次通解系数陷阱**：
    $x = k\xi + \eta$。其中 $k$ 是任意常数，但 $\eta$ 前面**没有系数**（或系数为1）。千万别写成 $C_1 \xi + C_2 \eta$。

## 第五章 特征值与特征向量 (Eigenvalues)

### 1. 定义与求解机制

*   **定义式**：$A\alpha = \lambda \alpha$ （其中 **$\alpha \neq 0$**）。
*   **求解步骤**：
    1.  解特征方程 $|\lambda E - A| = 0$，求出特征值 $\lambda_i$。
    2.  对每一个 $\lambda_i$，解齐次方程 $(\lambda_i E - A)x = 0$，求出基础解系，即为特征向量。

### 2. 核心性质（三位一体）

这是做选择填空题的秒杀工具，**必须烂熟于心**：

1.  **迹 (Trace)**：$\sum_{i=1}^n \lambda_i = \sum_{i=1}^n a_{ii} = \text{tr}(A)$。
2.  **行列式**：$\prod_{i=1}^n \lambda_i = |A|$。
3.  **不变性**：若 $A \sim B$（相似），则 $|A|=|B|$，$\text{tr}(A)=\text{tr}(B)$，特征值相同。

*   **函数性质**：
    若 $A$ 的特征值为 $\lambda$，则：
    *   $A^k \to \lambda^k$
    *   $A^{-1} \to 1/\lambda$
    *   $f(A) \to f(\lambda)$
    *   $A^* \to \frac{|A|}{\lambda}$ （若 $\lambda \neq 0$）

### 3. 相似对角化 (Diagonalization) —— 大题考点 No.1

判定矩阵 $A$ 是否能与对角阵 $\Lambda$ 相似 ($P^{-1}AP = \Lambda$)：

*   **充分条件**：
    *   $A$ 有 $n$ 个**互不相等**的特征值 $\implies$ 必可对角化。
    *   $A$ 是**实对称矩阵** ($A^T=A$) $\implies$ 必可对角化。

*   **充要条件**：
    $A$ 有 $n$ 个**线性无关**的特征向量。
    $\iff$ 对每个 $k$ 重特征值 $\lambda_i$，其**几何重数**等于**代数重数**。
    $$ n - r(\lambda_i E - A) = k $$
    *(解释：$k$ 是 $\lambda_i$ 作为根出现的次数，公式左边是对应齐次方程的基础解系个数)*

*   **反求矩阵 A**：
    若已知特征值 $\Lambda$ 和特征向量 $P=(\alpha_1, \dots, \alpha_n)$，则 **$A = P \Lambda P^{-1}$**。

## 第六章 二次型 (Quadratic Forms)

二次型本质上是对称矩阵的“影分身”。

### 1. 标准化 (Standardization)

将 $f(x) = x^T A x$ 化为标准形 $y^T \Lambda y = k_1 y_1^2 + \dots + k_n y_n^2$。

#### 方法一：正交变换法 (数一重点)
*   **原理**：$x = Qy$，其中 $Q$ 是正交矩阵 ($Q^T = Q^{-1}$)。
*   **步骤**：
    1.  求 $A$ 的特征值 $\lambda_1, \dots, \lambda_n$。
    2.  求特征向量，并进行 **Schmidt 正交化** 和 **单位化**，得到 $\gamma_1, \dots, \gamma_n$。
    3.  令 $Q = (\gamma_1, \dots, \gamma_n)$，则 $Q^T A Q = \Lambda = \text{diag}(\lambda_1, \dots, \lambda_n)$。
*   **几何意义**：保持向量长度不变，旋转坐标轴。标准形系数**就是特征值**。
*   **应用场景**：求二次型在 $x^T x = 1$ (单位球) 上的最大/最小值（即最大/最小特征值）。

#### 方法二：配方法 (Congruence)
*   **原理**：$x = Cy$，其中 $C$ 仅需可逆。
*   **步骤**：初高中配方技巧。
*   **注意**：标准形的系数 $k_i$ **不等于** 特征值，但正系数个数和负系数个数与特征值分布一致（**惯性定理**）。

### 2. 矩阵的合同 (Congruence)
*   **定义**：$C^T A C = B$，称 $A$ 与 $B$ 合同。
*   **判别**：实对称矩阵 $A, B$ 合同 $\iff$ 它们有相同的正惯性指数 $p$ 和负惯性指数 $q$。

### 3. 正定性判定 (Positive Definite)

判断 $f(x) = x^T A x > 0$ ($\forall x \neq 0$)，即矩阵 $A$ 是否正定：

*   **判据** (任意满足其一即可)：
    1.  **特征值**：全大于 0 ($\lambda_i > 0$)。
    2.  **顺序主子式**：全大于 0。
    3.  **惯性指数**：正惯性指数 $p = n$。
    4.  **合同**：$A$ 合同于 $E$ ($C^T A C = E$)。
    5.  **定义结构**：存在可逆阵 $Q$ 使得 $A = Q^T Q$。

## 补充：数一特有考点 (Vector Space)

*虽然考频低，但考出来就是送命题，需防守。*

1.  **基变换公式**：
    若旧基 $(\alpha_I) = (\alpha_1, \dots, \alpha_n)$，新基 $(\beta_I) = (\beta_1, \dots, \beta_n)$，且 $(\beta_I) = (\alpha_I)C$。
    则 $C$ 为**过渡矩阵**。

2.  **坐标变换公式** (反直觉！)：
    若向量 $\gamma$ 在旧基下坐标为 $x$，在新基下坐标为 $y$（即 $\gamma = (\alpha_I)x = (\beta_I)y$）。
    则坐标变换公式为：**$x = Cy$** 或 **$y = C^{-1}x$**。
    *(记忆：基变换是右乘，坐标变换是左乘且互逆)*

### 第五、六章 典型“母题”演练

#### 【母题 4】 相似对角化的参数讨论 (大题必练)
设 $A = \begin{pmatrix} 1 & 1 & a \\ 0 & 2 & 0 \\ 1 & 1 & 2 \end{pmatrix}$。
(1) 求 $a$ 的值，使 $A$ 可相似对角化。
(2) 求可逆矩阵 $P$，使 $P^{-1}AP$ 为对角阵。

> **思路**：
> 1.  先算特征值 $|\lambda E - A| = (\lambda-2)[\dots]$。
> 2.  若特征值互异 $\to$ 必可对角化。
> 3.  若有重根（例如 $\lambda_1 = \lambda_2 = 2$），则代入 $2E-A$，要求其秩 $r(2E-A) = n - 2 = 1$。由此定 $a$。

#### 【母题 5】 实对称矩阵的反求 (逆向思维)
设 3 阶实对称矩阵 $A$ 的特征值为 $\lambda_1=1, \lambda_2=2, \lambda_3=-2$，且 $\alpha_1 = (1, -1, 1)^T$ 是 $\lambda_1$ 的特征向量，求 $A$。

> **思路**：
> 1.  实对称矩阵不同特征值对应的特征向量**必正交**。
> 2.  设 $\alpha_2 = (x_1, x_2, x_3)^T$，则 $\alpha_2 \perp \alpha_1$。这通常条件不够。
> 3.  **更优解法**：利用谱分解定理 $A = \lambda_1 p_1 p_1^T + \lambda_2 p_2 p_2^T + \dots$ (需单位化)。
> 4.  或者：若题目给出 $\alpha_2, \alpha_3$ 的部分信息，利用正交性解出完整向量，再用 $A = P \Lambda P^{-1}$。**注意 $P$ 若是正交阵，则 $P^{-1}=P^T$，计算量大减。**

#### 【母题 6】 二次型参数与正定性
已知二次型 $f = x_1^2 + 4x_2^2 + 2x_3^2 + 2tx_1x_2$ 正定，求 $t$ 的范围。

> **思路**：
> 写出矩阵 $A = \begin{pmatrix} 1 & t & 0 \\ t & 4 & 0 \\ 0 & 0 & 2 \end{pmatrix}$。
> 利用顺序主子式全大于 0：
> 1阶：$1 > 0$ (ok)
> 2阶：$4 - t^2 > 0 \implies -2 < t < 2$
> 3阶：$2(4-t^2) > 0$
> 综上：$t \in (-2, 2)$。

### 补充一：向量空间与基变换 (对应考纲“三、向量” 第5-6点)
*这是数学一独有的内容，数学二、三不考，因此也是命题人用来区分数一考生的“杀手锏”。*

#### 1. 概念辨析
*   **向量空间**：解空间就是一种向量空间。
*   **维数**：极大线性无关组中向量的个数（对于 $Ax=0$，维数 $= n - r(A)$）。
*   **子空间**：比如 $Ax=0$ 的解集是 $\mathbb{R}^n$ 的子空间，但 $Ax=b (b\neq0)$ 的解集**不是**子空间（因为不含零向量，对加法不封闭）。

#### 2. 过渡矩阵与坐标变换 (高危易错点)
设 $V$ 中有两组基：旧基 $(\alpha) = (\alpha_1, \dots, \alpha_n)$ 和 新基 $(\beta) = (\beta_1, \dots, \beta_n)$。

*   **定义公式（基变换）**：
    $$ (\beta_1, \dots, \beta_n) = (\alpha_1, \dots, \alpha_n) \cdot C $$
    这里的 $C$ 称为 **由基 $\alpha$ 到基 $\beta$ 的过渡矩阵**。
    *(注意：$C$ 放在右边，且是列向量的形式)*

*   **坐标变换公式（反向操作）**：
    设向量 $\gamma$ 在旧基 $\alpha$ 下的坐标为 $x$（即 $\gamma = (\alpha)x$），在新基 $\beta$ 下的坐标为 $y$（即 $\gamma = (\beta)y$）。
    则坐标关系为：
    $$ x = Cy \quad \text{或} \quad y = C^{-1}x $$
    > **口诀**：基变换 $C$ 在右，坐标变换 $C$ 在左；求新坐标 $y$ 要用 $C$ 的逆。

### 补充二：反对称矩阵 (对应考纲“二、矩阵” 第1点)
*考纲明确提到了“反对称矩阵”，这是之前的笔记里未详细展开的特殊矩阵。*

*   **定义**：$A^T = -A$。
*   **必背性质**：
    1.  **主对角线元素必全为 0** ($a_{ii} = 0$)。
    2.  **行列式性质**：
        *   若 $n$ 为 **奇数**，则 $|A| = 0$。（证明：$|A| = |-A^T| = (-1)^n |A| = -|A| \implies |A|=0$）。
        *   若 $n$ 为 **偶数**，则 $|A|$ 是某个多项式的平方，必 $\ge 0$。
    3.  **特征值**：实反对称矩阵的特征值只能是 **0 或 纯虚数**。
        *(考研主要考实数域，所以如果出题，通常特征值为 0，或者题目会引导你利用 $A^T=-A$ 进行代数变形)*

---

### 补充三：二次型的规范形 (对应考纲“六、二次型” 第1点)
*之前的笔记重点讲了“标准形”，考纲里多了一个“规范形”。*

*   **标准形**：$d_1 y_1^2 + \dots + d_n y_n^2$ （系数 $d_i$ 只要不为0即可，通常是特征值）。
*   **规范形**：系数只能是 **$1, -1, 0$**。
    $$ z_1^2 + \dots + z_p^2 - z_{p+1}^2 - \dots - z_{p+q}^2 $$
*   **惯性定理**：无论选取什么坐标变换（只要可逆），化为标准形后，**正系数的个数 $p$**（正惯性指数）和 **负系数的个数 $q$**（负惯性指数）是**唯一确定**的。
*   **考点**：两个矩阵合同 $\iff$ 它们具有相同的正惯性指数 $p$ 和负惯性指数 $q$ $\iff$ 它们的规范形完全一致。

---

### 补充四：克拉默法则 (Cramer's Rule) (对应考纲“四、线性方程组” 第1点)
*虽然实际计算中很少用，但理论题或填空题可能会涉及。*

*   **适用条件**：
    1.  方程个数 = 未知数个数 ($n$ 个方程，$n$ 个未知数)。
    2.  系数行列式 **$D = |A| \neq 0$**。
*   **结论**：方程组有**唯一解**。
    $$ x_j = \frac{D_j}{D} $$
    其中 $D_j$ 是把 $D$ 中第 $j$ 列替换为常数项列 $b$ 后的行列式。
*   **避坑**：如果 $D=0$，克拉默法则失效，此时方程组可能无解或有无穷多解（需用秩判定）。

---

### 补充五：解空间的概念 (对应考纲“四、线性方程组” 第3点)
*   对于齐次方程组 $Ax=0$，所有解构成的集合 $S = \{x | Ax=0\}$ 是一个向量空间，称为**解空间**。
*   **解空间的基** = 基础解系。
*   **解空间的维数** = $n - r(A)$。
*   此概念常用于将“求基础解系”的问题包装成“求解空间的基与维数”。
